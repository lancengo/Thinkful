```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
import statsmodels.api as sm
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

postgres_user = 'dsbc_student'
postgres_pw = '7*.8G9QH21'
postgres_host = '142.93.121.174'
postgres_port = '5432'
postgres_db = 'weatherinszeged'

engine = create_engine('postgresql://{}:{}@{}:{}/{}'.format(
    postgres_user, postgres_pw, postgres_host, postgres_port, postgres_db))
df = pd.read_sql_query('select * from weatherinszeged',con=engine)

engine.dispose()
```

```python
X = df[['humidity', 'windspeed']]
Y = df['apparenttemperature'] - df['temperature']
X = sm.add_constant(X)

results = sm.OLS(Y, X).fit()

results.summary()
```

Yes both coefficients are statistically significant with a p-value of 0. When the target variable is positive, it means that the perceived temperature is hotter than it actually is. Also when the target variable is negative, then the perceived temperature is colder than it actually is. Both coefficients are negative. This means that as humidity increases, people think that it's colder than it is. Also when windspeed increases, people think it is colder than it is. the wind speed makes sense as more wind increases wind chill. the humidity can cause people to think that the temperature is colder or hotter depending on the temperature. a higher humidity makes people think it is colder than a higher windspeed, according to the coefficients. 

```python
df['humid_wind'] = df.windspeed * df.humidity

X = df[['humidity', 'windspeed', 'humid_wind']]
Y = df['apparenttemperature'] - df['temperature']
X = sm.add_constant(X)

results = sm.OLS(Y, X).fit()

results.summary()
```

the coefficients are all statistically significant. the signs of humidity and windspeed are positive while humidity_wind interaction is negative. the humidity_wind interaction is the largest and shows that it affects the target variable the most.
